## Histograma

Histogramas recebem apenas uma variável. O eixo Y é a quantifiação das incidências dentro de certas faixas

## Gráfico de Setores (Pizza)

```python
# O gráfico é desenhado seguindo a proporção de incidências.
x = [11,19,27,42]

# Títulos dos setores
itens = ["Maçãs", "Bananas", "Cerejas", "Peras"]

# Por padrão, o ângulo se inicia no 0°.
ang_inicial = 30

# "explode" separa certos setores do eixo
lacuna = [0.2,0,0.1,0]

# É possível determinar a cor de cada setor, ao invés de deixar as cores padrão
cores = ["black", "hotpink", "b", "#4CAF50"]

# "shadow" cria sombreamento

plt.pie(x, labels = itens, startangle = ang_inicial, explode = lacuna, colors = cores, shadow = True)

# legendas podem ser usadas como alternativa aos nomes
plt.legend(title = "Frutas")
```

## Subplot

```python
xA = [0, 1, 2, 3]
yA = [3, 8, 1, 10]

xB = [1, 2, 3, 4, 5, 6]
yB = [10, 20, 50, 10, 30, 40]
```

Função que permite desenhar múltiplos gráficos em uma mesma imagem.

```python
plt.subplot(2,2,1) # A imagem tem duas linhas e duas colunas, e o próximo gráfico é o 1º
plt.plot(xA,yA)

plt.subplot(2,2,4) # A imagem tem duas linhas e duas colunas, e o próximo gráfico é o 4º
plt.bar(xB,yB)
```

```python
# Subplots são como vetores de gráficos e atendem à mesma indexação

fig, ax = plt.subplots(2,2)

ax[0,0].plot(xA,yA)

ax[1,1].bar(xB,yB)
```

```python
# Rótulos ficam abaixo da definição dos valores e acima do próximo comando de subplot.

plt.subplot(2,1,1)
plt.plot(xA,yA)
plt.title('Gráfico A')

plt.subplot(2,1,2)
plt.bar(xB,yB)
plt.title('Gráfico B')

# Um título pode ser adicionado à imagem

plt.suptitle("Gráficos", size = 20)
```

```python
# Espaço entre os gráficos

plt.subplot(2,1,1)
plt.plot(xA,yA)
plt.title('Gráfico A')

plt.subplot(2,1,2)
plt.bar(xB,yB)
plt.title('Gráfico B')

plt.suptitle("Gráficos", size = 20)

# Os valores são percentuais da imagem

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.86,
                    wspace=0.8,
                    hspace=0.8)
```

### Gráfico de Linha

```python
# Criar uma área de plotagem
figura = plt.figure()

# Criar eixos
 #  Todo gráfico é feito dentro dos eixos
 #  O parâmetro é uma lista com as coordenadas das pontas dos retângulos
eixos = figura.add_axes([0,0,0.8,0.8])

# Criar plot
 #  O método 'plot' evoca o gráfico de linha
 #  O terceiro parâmtro define a cor da linha
eixos.plot(x,y,'r')

# Rótulos e títulos
eixos.set_xlabel('x')
eixos.set_ylabel('y')
eixos.set_title('Gráfico de Linha')
```

```python
# Criando dois gráficos

fig = plt.figure()

# Dois eixos
 #  [distância X do início, distância Y do início, tamanho X, tamanho Y]
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

# Principal
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Gráfico Maior')

# Secundário
axes2.plot(y,x,'g')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Gráfico Menor')
```

```python
# Gráficos Paralelos

# O resultado do comando 'subplots' é uma tupla contendo dois elementos:
  # 1 > figura do matplotlib
  # 2 > ndarray do numpy, contendo X outros ndarray, sendo X o numero de linhas (nrows)
    # a) cada ndarray relacionado às linhas contém um 'matplotlib.axes._subplots.AxesSubplot'
fig, axes = plt.subplots(nrows = 1, ncols = 9)

for ax in axes:
  ax.plot(x,y,'r')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('Gráfico Maior')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3)

for ax in axes:
  for a in ax:
    a.plot(x,y,'r')
    a.set_xlabel('x')
    a.set_ylabel('y')
    a.set_title('Título')
```

```python
# set_xscale; set_yscale: Alteração de escala

fig, axes = plt.subplots(1,2,figsize = (10,4))

axes[0].plot(x,y,x,exp(x))
axes[0].set_title("Escala padrão")

axes[1].plot(x,y,x,exp(x))
axes[1].set_yscale('log')       #<<<
axes[1].set_title("Escala logarítmica")

# {"linear", "log", "symlog", "logit", ...}
```

```python
# Adicionando grade

fig, axes = plt.subplots(1,2,figsize = (15,4))

axes[0].plot(x,y,x,exp(x), lw = 2)
axes[0].grid(True)                             #<<<

axes[1].plot(x,y,x,exp(x), lw = 2)
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)       #<<<
```

### Outros tipos de gráficos

```python
fig, axes = plt.subplots(2,2,figsize = (15,8))

axes[0][0].scatter(x,y,x,exp(x))
axes[0][0].set_title("Scatter Plot")

axes[0][1].step(x,y,x,exp(x))
axes[0][1].set_title("Escada")

axes[1][0].bar(x,y,x,exp(x))
axes[1][0].set_title("Barra")

axes[1][1].fill_between(x,y,x,exp(x))
axes[1][1].set_title("Linha Preenchida")
```

# Gráfico 3D

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
alpha = 0.7
phi_ext = 2*np.pi*0.5

print(alpha)
print(phi_ext)

print()

def ColorMap(phi_m, phi_p):
    return (+alpha-2*np.cos(phi_p)*np.cos(phi_m)-alpha*np.cos(phi_ext-2*phi_p))

# Dado
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p,phi_m)
Z = ColorMap(X,Y).T

print(X)
print(Y)
print(Z)
```

```python
# Outra maneira de criar subplots é com o comando '.add_subplot'.

fig = plt.figure(figsize = (14,10))

# Gráfico 1
ax = fig.add_subplot(2,2,1, projection = '3d') # de uma grade 1x2, este é o gráfico 1
p = ax.plot_surface(X,Y,Z, rstride = 4, cstride = 4, linewidth = 0)

# Gráfico 2
ax = fig.add_subplot(2,2,2, projection = '3d') # de uma grade 1x2, este é o gráfico 2
p = ax.plot_surface(X,Y,Z, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.coolwarm, antialiased = False)
# Legenda
cb = fig.colorbar(p, shrink = 0.5)

# Gráfico 3
ax = fig.add_subplot(2,2,3, projection = '3d')
p = ax.plot_wireframe(X,Y,Z, rstride = 4, cstride = 4)

```
