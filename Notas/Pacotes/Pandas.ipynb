## Séries

Séries são como colunas de uma tabela, são arrays unidimensionais.

```python
Lista = [1,2,3,4,5,6,7,8,9]

Série1 = pandas.Series(Lista)

print(Série1)
print(type(Série1))

# O índice para cada valor é criado automaticamente.
```

```python
# Outras informações sobre as séries

print(Série1.values)

print(Série1.index)
```

```python
print(Série1.info())
```

```python
# Criando uma série e definindo o índice

Lista = [1,2,3,4,5,6,7,8,9]
Índices = ['a','b','c','d','e','f','g','h','i']

Série2 = pandas.Series(Lista, index = Índices)

print(Série2)
print(type(Série2))

print(Série2.index)
```

```python
# Criação de série a partir de um dicionário

Dicionário = {'a':1,'b':2,'c':3,'d':4,'e':5,'f':6,'g':7,'h':8,'i':9}

Série3 = pandas.Series(Dicionário)


print(Série3)
print(type(Série3))
```

```python
# Para um mesmo índice, um mesmo valor

Outro_Índice = ['a','a','a','a','a','a','a','a','a','a']

Série4 = pandas.Series(Dicionário, index = Outro_Índice)


print(Série4,'\n')

# Mudar o dicionário não muda a série. Eles estão desvinculados

Dicionário['a'] = 2

print(Série4,'\n')

# Para novos índices, não há valor

Outro_Índice = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n']

Série4 = pandas.Series(Dicionário, index = Outro_Índice)


print(Série4,'\n')
```

```python
# Slicing

Série2[Série2>3]
```

```python
# Valor de certo índice

Série2['d']
```

```python
# resposta booleana para verificação de continência

print('d' in Série2)

print('d' in Série2.index)

print(3 in Série2)

print(3 in Série2.values)
```

```python
# Verificação de valores nulos

print(pandas.isnull(Série4),'\n')

print(Série4.isnull())
```

```python
# Verificação de valores não-nulos

print(pandas.notnull(Série4),'\n')

print(Série4.notnull())
```

```python
# Verificação condicional

print(Série4 > 5)
```

```python
# Soma/concatenação entre séries

print(Série3 + Série4)
```

```python
# Valores NaN permanecem dessa forma depois de passar por uma operação matemática.

print(Série1 + Série3)
```

```python
# Determinando a característica "nome".

Série4.name = 'Exemplo' # nome do objeto

Série4.index.name = 'Letra' # Nome do índice

print(Série4)
```

### Séries Temporais

Séries cujos índices estão em formato temporal - consequentemente, cada registro corresponde a um momento.

```python
N = 10

Datas = pandas.date_range('31/12/1989',periods=N) # N índices temporais

print(type(Datas),'\n')

from numpy.random import randint

RNG = randint(0,500,N) # N números aleatórios, de 0 a 499

Série_Temporal = pandas.Series(RNG,Datas)
print(Série_Temporal)
```

## Alterações de Valores

A função map() recebe um dicionário como parâmetro que serve de instrução sobre como alterar valores de uma variável originalmente categórica.

```python
Instructions = {'Norte':0,'Nordeste':1,"Centro Oeste":2,'Sudeste':3,'Sul':4}

Tabela2['Região'].map(Instructions)
```

A função *.loc[ ]*, abordada anteriormente como uma forma de fatiamento horizontal, recebendo verificações lógicas como índice, permite substituir uma gama de valores de forma específica.  

Os operadores lógicos não são os convencionalmente utilizados na programação em python:  
* and - &
*  or - |  

Se duas ou mais verificações lógicas forem necessárias, é preciso separar cada uma com parêntesis.

```python
print(Tabela1,'\n')

Tabela_Loc = Tabela1.copy()

# DataFrame.loc(Verificação lógia, coluna modificada/criada) = valor
Tabela_Loc.loc[(Tabela_Loc['PIB'] > Tabela_Loc['PIB'].mean()) & (Tabela_Loc['População'] > Tabela_Loc['População'].mean()), 'Lista'] = 100

print(Tabela_Loc)
```
